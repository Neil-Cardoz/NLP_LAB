```python
import os
import sys
import transformers #library for pre-trained NLP models
import tensorflow as tf
from datasets import load_dataset
from transformers import AutoTokenizer #Automatically fetches the correct tokenizer for a given pre-trained model.
from transformers import TFAutoModelForSeq2SeqLM, DataCollatorForSeq2Seq #Loads a TensorFlow version of a sequence-to-sequence model | Helps prepare batches for seq2seq training.
from transformers import AdamWeightDecay # optimizer (Adam with weight decay)
```

```python
model_checkpoint = "Helsinki-NLP/opus-mt-en-hi"
```

```python
raw_datasets['train'][1]
```

```python
tokenizer = AutoTokenizer.from_pretrained(model_checkpoint)

# AutoTokenizer → Hugging Face class that automatically picks the right tokenizer type for the given model.
# .from_pretrained(model_checkpoint) → Downloads and loads the pre-trained tokenizer associated with "Helsinki-NLP/opus-mt-en-hi".
# tokenizer → Variable now holds the tokenizer, which can:
# Convert English/Hindi text into token IDs (numbers) that the model understands.
# Convert token IDs back into human-readable text after prediction.
```

```python
tokenizer("Hello, this is a sentence!")

# nput_ids = encoded version of the sentence.
# attention_mask tells the model which positions to focus on.
# attention_mask → 1s for real tokens, 0s for padding.
```

```python
tokenizer(["Hello, this is a sentence!", "This is another sentence."])
```

```python
with tokenizer.as_target_tokenizer():  #This tells Hugging Face that the tokenizer should work in target language mode (Hindi here)
    print(tokenizer(["एक्सेर्साइसर पहुंचनीयता अन्वेषक"]))
```

```python
max_input_length = 128
max_target_length = 128

source_lang = "en"
target_lang = "hi"


def preprocess_function(examples):
    inputs = [ex[source_lang] for ex in examples["translation"]]  # Extracts English sentences from the dataset
    targets = [ex[target_lang] for ex in examples["translation"]]  # Extracts Hindi sentences from the dataset
    model_inputs = tokenizer(inputs, max_length=max_input_length, truncation=True) # Tokenizes English sentences into input_ids & attention_mask, truncating if longer than 128 tokens.

    # Setup the tokenizer for targets
    with tokenizer.as_target_tokenizer():
        labels = tokenizer(targets, max_length=max_target_length, truncation=True)   # Switch tokenizer into target mode and tokenize Hindi sentences

    model_inputs["labels"] = labels["input_ids"] #Attach Hindi token IDs as "labels" so the model learns to predict them from English input.

    return model_inputs


  #This function takes raw English–Hindi sentence pairs and converts them into numerical input-output pairs ready for a seq2seq translation model.
```

```python
preprocess_function(raw_datasets["train"][:2])

#two parallel English–Hindi sentences are converted into numerical tensors that the translation model can train on.
```

```python
tokenized_datasets = raw_datasets.map(preprocess_function, batched=True)
```

```python
model = TFAutoModelForSeq2SeqLM.from_pretrained(model_checkpoint)

# TFAutoModelForSeq2SeqLM → Hugging Face class that loads a TensorFlow sequence-to-sequence model (encoder–decoder), useful for machine translation, summarization, etc.
# .from_pretrained(model_checkpoint) → Downloads the pre-trained model weights & config from the Hugging Face Hub.
# model → Now holds the ready-to-use translation model that can be fine-tuned or used directly for inference.
```

```python
batch_size = 16       # Number of examples processed in one forward/backward pass before updating model weights.
learning_rate = 2e-5  # Step size for the optimizer; controls how much to adjust weights on each update
weight_decay = 0.01   # A form of regularization: prevents overfitting by penalizing large weights (shrinks weights slightly toward zero)
num_train_epochs = 1  # Number of complete passes over the training dataset
```

```python
data_collator = DataCollatorForSeq2Seq(tokenizer, model=model, return_tensors="tf")  #Data collator passes the data to the model in batches
```

```python
generation_data_collator = DataCollatorForSeq2Seq(tokenizer, model=model, return_tensors="tf", pad_to_multiple_of=128)

#generation_data_collator automatically batches and pads input/label sequences (to multiples of 128)
# so the translation model can run efficiently during evaluation or text generation.
```

```python
train_dataset = model.prepare_tf_dataset(
    tokenized_datasets["test"],
    batch_size=batch_size,
    shuffle=True,
    collate_fn=data_collator,
)

# This creates a TensorFlow dataset of padded, shuffled mini-batches (size 16) from your test split, ready to be fed into the model.
```

```python
validation_dataset = model.prepare_tf_dataset(
    tokenized_datasets["validation"],
    batch_size=batch_size,
    shuffle=False,
    collate_fn=data_collator,
)

# This creates a non-shuffled, padded TensorFlow dataset from the validation split, used to evaluate the model’s performance during/after training.
```

```python
generation_dataset = model.prepare_tf_dataset(
    tokenized_datasets["validation"],
    batch_size=8,
    shuffle=False,
    collate_fn=generation_data_collator,
)

#This creates a TensorFlow dataset of validation samples (batch size 8, padded for generation) that the model will use to generate translations for evaluation.
```

```python
optimizer = AdamWeightDecay(learning_rate=learning_rate, weight_decay_rate=weight_decay)
model.compile(optimizer=optimizer)

# This configures the model with an AdamW optimizer (with learning rate + weight decay) and makes it ready for training.
```

```python
model.fit(train_dataset, validation_data=validation_dataset, epochs=1)

# This trains your English→Hindi translation model for 1 full pass over the training dataset, while monitoring performance on the validation set.
```

```python
model.save_pretrained("tf_model/")

# It saves your fine-tuned model in Hugging Face’s pretrained model format inside the folder tf_model/.
```
